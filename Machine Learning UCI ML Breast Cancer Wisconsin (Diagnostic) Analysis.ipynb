```python
#Importing all the packages like pandas, numpy, seaborn, sklearn,
import pandas as pd
import numpy as np
import seaborn as sns; sns.set(style="ticks", color_codes=True)
from sklearn.datasets import load_breast_cancer

#Impoting in matplitlib and setting the requirement for plot to populate in the notebook
import matplotlib.pyplot as plt
%matplotlib inline

#Calling in the Dataset, describing the dataset and storing it as a Data frame. 
cancer = load_breast_cancer()
print(cancer.DESCR)
df = pd.DataFrame(np.c_[cancer['data'], cancer['target']],
                  columns= np.append(cancer['feature_names'], ['target']))
```

```python
#Seperating the features and target values into X and y vectors
X = df.drop('target', axis=1).values
y = df['target'].values
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
def my_pipeline(test_size =.3, random_state =0):
    
    """
    The function has the following operations and steps: 
    1. Divide the data into training and testing sets 
    2. Declare the Classifer I want to use 
    3. Train the Trainging set on the training data 
    3. Test the model on my testing data 
    56. Pl
    """
    
    #Dividing the data into training and testing sets
    #Randomly choosing 30% of the data fro testing 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
    print(X_test.shape)
    
    
    #Declaring the Classifier I want to use 
    myclassifier = KNeighborsClassifier()
    
    #Train the Trainging set on the training data 
    myclassifier.fit(X_train, y_train) 
    
    #Using the model to Pmake predcition on th test set. 
    predictions = myclassifier.predict(X_test)
    print (accuracy_score(y_test, predictions))
    
    
    # Creating two list for the trainging accuracy and testing accuracry respectively 
    training_accuracy = []
    test_accuracy = []

    # Setting a range for testing for the 1-10 neighbors
    neighbors_settings = range(1,11)

    # iterate KNN through the different number of neighbors to determine the most appropriate (best)
    for n_neighbors in neighbors_settings:
        clf = KNeighborsClassifier(n_neighbors=n_neighbors)
        clf.fit(X_train, y_train)
        training_accuracy.append(clf.score(X_train, y_train))
        test_accuracy.append(clf.score(X_test, y_test))

    # Plotting the Results and comparing the accuracies
    plt.plot(neighbors_settings, training_accuracy, label='Accuracy of the training set')
    plt.plot(neighbors_settings, test_accuracy, label='Accuracy of the test set')
    plt.ylabel('Accuracy')
    plt.xlabel('Number of Neighbors')
    plt.legend()
```

```python
#Running my Pipeline Function
my_pipeline()
```
