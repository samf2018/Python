##  1) What is the popular Cosmetic companies ‘Type’ of content that influences active engagement (Likes, Comments,Shares) among users? 

```python
Type_mean = facebook_data_new.groupby('Type').mean() # group engagement by type

fig, axs = plt.subplots(1,3,figsize=(13,4)) # allow all three graphs to be displayed 

Type_mean.Likes.plot('bar', title='Average # of likes', ax=axs[0])
Type_mean.Comments.plot('bar', title='Average # of Comments' , ax=axs[1])
Type_mean.Shares.plot('bar', title='Average # of shares', ax = axs[2])

#Link has the least likes, Comments, and Shares
#Video has the most Likes, Comments, and Shares
#Photo has more likes than Status, but Status gets more comments and shares than Photo
```

### Which from Type has more reach and or impressions ?

```python
# Comparing Type with Lifetime Post Total Reach and Lifetime Post Total Impressions
fig, axs = plt.subplots(1,2,figsize=(16,5))

Type_mean['Lifetime Post Total Reach'].plot('bar', title='Reach', ax=axs[0])
Type_mean['Lifetime Post Total Impressions'].plot('bar', title='Impressions', ax=axs[1])
```

##  2) Which month has the highest engagement among users? 

```python
# Best Month
# (Lifetime Post Consumers, Lifetime Post Consumptions and Total Interactions)
facebook_data_new_month = facebook_data_new.groupby('Post Month').mean()

ax = facebook_data_new_month['Lifetime Post Consumers'].plot('line', legend='Lifetime Post Consumers', 
                                                             title = 'User Engagment With Company Post by Month')
facebook_data_new_month['Lifetime Post Consumptions'].plot('line', legend='Lifetime Post Consumptions', ax=ax)
facebook_data_new_month['Total Interactions'].plot('line', legend='Lifetime Engaged Users', ax=ax)

#Our graph shows us the following information:
#There is a large peak in the month of February which shows that users are engaged during this month by clicking and interacting with posts.
#Also, we see June through July were the times when Lifetime Post Consumers, Consumptions and Total Interactions are all peacking before falling back down after July. 

```

*  What is the best month specifically for the most active engagements (Likes, Comments and Shares) from users?

```python
# Best Month
#  Best month specifically for the most active engagements(Likes, Comment, Shares) from users
facebook_data_new_month_mean = facebook_data_new.groupby('Post Month').mean()
facebook_data_new_month_sum = facebook_data_new.groupby('Post Month').sum()
fig, axs = plt.subplots(1,2,figsize=(16,4))

facebook_data_new_month_mean['Likes'].plot('line', legend='Likes', title = 'Average Interaction by Month (Mean)' , ax= axs[0])
facebook_data_new_month_mean['Comments'].plot('line', legend='Comments', ax=axs[0])
facebook_data_new_month_mean['Shares'].plot('line', legend='Shares', ax=axs[0])


facebook_data_new_month_sum['Likes'].plot('line', legend='Lifetime Post Consumers', 
                                            title = 'Total Interaction by Month (Sum)' , ax= axs[1])
facebook_data_new_month_sum['Comments'].plot('line', legend='Lifetime Post Consumptions', ax=axs[1])
facebook_data_new_month_sum['Shares'].plot('line', legend='Lifetime Engaged Users', ax=axs[1])

#We see that calculating by both mean and sum we find that it is showing the same trend
# In other words, the most active engagements (Likes, Commenetes and Shares) are happening between the months of June and August.
#We also see that there are peaks noticeably in the month of February and April as well
```

##  3) Which month has the highest reach and impressions among users? (Reach & Impressions)

```python
fig, axs = plt.subplots(1,2,figsize=(16,4))
facebook_data_new_month_mean['Lifetime Post Total Reach'].plot('line', legend='Likes', 
                                                               title = 'Average of Reach and Impressions (Mean)' , 
                                                               ax= axs[0])
facebook_data_new_month_mean['Lifetime Post Total Impressions'].plot('line', legend='Comments', ax=axs[0])

facebook_data_new_month_sum['Lifetime Post Total Reach'].plot('line', legend='Reach', 
                                                              title = 'Total of Reach and Impressions (Sum)' , 
                                                              ax= axs[1])
facebook_data_new_month_sum['Lifetime Post Total Impressions'].plot('line', legend='Impressions', ax=axs[1])
#We see that February has the most Impression up to 80,000 where has July was the best month for the most reach.
#We also see that the general trend is still the same in how the peaks rise and fall around the same month.
```

##  4) Which day out of the week has the highest active engagement (Likes, Shares and Comments) among users? 

```python
facebook_data_new_weekday = facebook_data_new.groupby('Post Weekday').mean()
fig, axs = plt.subplots(1,2,figsize=(16,4))

facebook_data_new_weekday['Likes'].plot('line', legend='Likes', 
                                        title = 'Average Interaction by Month (Mean)' , ax= axs[0])
facebook_data_new_weekday['Comments'].plot('line', legend='Comments', ax=axs[0])
facebook_data_new_weekday['Shares'].plot('line', legend='Shares', ax=axs[0])


facebook_data_new_weekday['Shares'].plot('bar', legend='Shares', ax=axs[1])
#We see that engagement increased from Tuesday to Wednesday and even Thursday.
#Thus, the best day seems to be Wednesday followed by Thursday and Fiday. 
# The least interaction happens on Sunday.
```

##  6) What is the best time of the day to get users actively engaged?

```python
facebook_data_hour = facebook_data_new.groupby('Post Hour').mean()
fig, axs = plt.subplots(1,2,figsize=(16,3))

facebook_data_hour['Likes'].plot('line', legend='Likes', title = 'Average Interaction per Hour (Mean)' , 
                                                             ax= axs[0])
facebook_data_hour['Comments'].plot('line', legend='Comments', ax=axs[0])
facebook_data_hour['Shares'].plot('line', legend='Shares', ax=axs[0])

facebook_data_hour['Shares'].plot('bar', legend='Shares', ax=axs[1])
#We see that active engagement peaks at 4-5am, 9-10am, 1-3pm, 3-5pm, 7-8pm
#The highest peak is 4-5am followed by 1-3pm or 13 to 15pm
#Important to note that these are the times people generally wake up, 
#take lunch breaks, and leave work depending on their location or timezone
```

##  7) What is the best time of the day where Reach and Impressions are high?

```python
ax = facebook_data_hour['Lifetime Post Total Reach'].plot('line', legend='Reach', 
                                                          title = 'Total of Reach and Impressions (Sum)' )
facebook_data_hour['Lifetime Post Total Impressions'].plot('line', legend='Impressions', ax=ax)
ax.figure.set_size_inches(16,6)

#IMPRESSIONS- Looking at this graph, we can see clearly that the largest peak or best time for Impressions are 
#happening at 6-7 followed by 13-14 hour as the second highest.

#REACH- As for Reach, the two highest peak or best time are at 5 and 14. Thus, this consmetic company should 
#take time into consideration especially since it's a metric that conveys the possibility of a post going viral
```

## 8) Does paid content influence engagement?

```python
fig, axs = plt.subplots(1,2,figsize=(16,4))

facebook_data_new.plot.scatter( x='Lifetime Post Total Reach', y= 'Lifetime Post Total Impressions',c='Paid',
                colormap ='viridis' ,
                title ='Reach and Impression of Users by Paid vs. Organic Content', ax = axs[0])


facebook_data_new.plot.scatter( x='Lifetime Post Total Reach', y= 'Comments',c='Paid',
                colormap ='viridis' ,
                title ='Engagment of Users by Paid vs. Organic Content', ax = axs[1])

#We see that there is not much difference with paid or organic content but we see that as the 
#Lifetime Post Total Reach and Lifetime Post Total Impression increase greater than 25,000 
#there seems to be both organic and paid content involved. 
#This tells us that this Cosmetic company can get high reach organically so its not manditory that they pay for 
#their content to be seen by current and new users
```

```python
#Remove Outlier for a better visual
top90_facebookdata = facebook_data_new.quantile(0.90)['Lifetime Post Total Reach'] 
                            # getting rid of outliers by taking 90% of our data  
facebook_data_new_modified = facebook_data_new[ facebook_data_new['Lifetime Post Total Reach'] 
                                                               < top95_facebookdata ]

fig, axs = plt.subplots(1,2,figsize=(16,4))

facebook_data_new_modified.plot.scatter( x='Lifetime Post Total Reach', y= 'Lifetime Post Total Impressions',
                                        c='Paid',
                                        colormap ='viridis' ,
                                        title ='Reach and Impression of Users by Paid vs. Organic Content', 
                                        ax = axs[0])

facebook_data_new_modified.plot.scatter( x='Lifetime Post Total Reach', y= 'Shares',c='Paid',
                colormap ='viridis' ,
                title ='Reach and Shares of Users by Paid vs. Organic Content', ax = axs[1])

# We got rid of the outliers in Reach and Impression of Users by Paid vs. Organic Content 
# Even so, we see one point where there is a very high Impression that went viral and was seen by new users
# and popular among users as seen by Reach.
# Also, we see a good pattern of both paid and organic posts that got shares with increasing reach which is 
# valuable information for the consmetic company because they see that they can create organic content that would 
# be shared without spending money
```

```python
fig, axs = plt.subplots(1,3,figsize=(16,4))

Paid_facebook_data_new = facebook_data_new.groupby('Paid').mean()

Paid_facebook_data_new['Lifetime Post Total Reach'].plot('bar', ax=axs[0], title = 'Reach Vs Paid ')
Paid_facebook_data_new['Lifetime Post Total Impressions'].plot('bar', ax=axs[1], title = 'Impressions Vs Paid ')
Paid_facebook_data_new['Total Interactions'].plot('bar', ax=axs[2], title = 'Interaction Vs Paid ')

# We see from this graph that Paid content has the most reach, Impressions and Interactions
# Thus, this cosmetic company can continue to do Paid content because they are getting a return 
# on there investments.
```
